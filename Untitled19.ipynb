```python
#step 1 :  import library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# step 2: load dataset
df = pd.read_csv("C:\\Users\\NMC\\Downloads\\train (1).csv")
```

```python
# first five rows 
df.head()
```

```python
# Missing values count
missing_before = df.isnull().sum()
print(missing_before)
```

```python
# step 3 : missing data patterns bar chart
missing_before[missing_before > 0].plot(kind='bar')
plt.title("Missing Values Before Cleaning")
plt.ylabel("Count")
plt.show()
```

```python
#step 4 : median imputation
# select numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].median())
```

```python
#step 5 : mode imputation
# select categorical columns
cat_cols = df.select_dtypes(include=['object']).columns

for col in cat_cols:
    if df[col].isnull().sum() > 0:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# step 6 : highly missing values drop # percentage of missing values
missing_percent = (df.isnull().sum() / len(df)) * 100

# columns with more than 50% missing
high_missing_cols = missing_percent[missing_percent > 50].index

# drop those columns
df.drop(columns=high_missing_cols, inplace=True)

print("Dropped columns:", list(high_missing_cols))
```

```python
# check missing values again
df.isnull().sum().sort_values(ascending=False).head()

# check dataset shape
df.shape
```

```python
#step 7 : comaparing before and after data set 
print("Before cleaning shape:", df.shape)
```

```python
print("After cleaning shape:", df.shape)
```

The House Price dataset was loaded and checked for missing values using Pandas.
Missing data patterns were visualized using bar charts.

Numerical columns were cleaned using **median imputation**, and categorical
columns were cleaned using **mode imputation**. Columns with very high missing
values were checked, but no columns were removed.

After cleaning, the dataset contained no missing values and its size remained
unchanged. The data is now clean and ready for further machine learning tasks.
